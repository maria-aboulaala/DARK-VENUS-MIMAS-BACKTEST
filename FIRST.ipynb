```python
import pandas as pd
```

```python
df = pd.read_csv("test1.csv")
df.head()
```

```python
df = pd.DataFrame(df)
```

```python
df.set_index('<DATE>', inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df
```

```python

df.index = df.index - pd.Timedelta(hours=2)
```

```python
import matplotlib.pyplot as plt



# Plot the 'balance' and 'equity' columns
plt.figure(figsize=(12, 6))  # Set the figure size (optional)

# Plot 'balance' in blue and 'equity' in green with labels
plt.plot(df.index, df['<BALANCE>'], label='Balance', color='blue')
plt.plot(df.index, df['<EQUITY>'], label='Equity', color='green')

# Add labels and a legend
plt.xlabel('Date')
plt.ylabel('Amount')
plt.title('Balance and Equity Over Time')
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()  # Ensures labels are not cut off when saving the plot
plt.show()
```

```python
import plotly.express as px

# Create a line plot using Plotly Express
fig = px.line(df, x=df.index, y=['<BALANCE>', '<EQUITY>'], labels={'index': 'Date', 'value': 'Amount'})

# Add a title
fig.update_layout(title='Balance and Equity Over Time')

# Show the plot
fig.show()
```
